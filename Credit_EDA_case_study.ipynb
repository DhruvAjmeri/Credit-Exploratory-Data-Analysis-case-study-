```python
def cat_uni1(app,x_axis,heading,color,hue=None):
    
    temp = pd.Series(data = hue)
    figure, graph = plt.subplots()
    col_width = len(app[x_axis].unique()) + 6 + 5*len(temp.unique())
    figure.set_size_inches(col_width , 10)
    plt.xticks(rotation=-45)
    plt.yscale('log')
    plt.title(heading)
    graph = sns.countplot(data = app, x= x_axis, order=app[x_axis].value_counts().index,hue = hue,palette=color) 
        
    plt.show()
```
